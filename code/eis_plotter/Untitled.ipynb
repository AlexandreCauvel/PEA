```python
import schemdraw
import schemdraw.elements as elm
from schemdraw.segments import Segment, SegmentText

class Warburg(elm.Element2Term):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        # Define ONLY the box body (width = 1.0)
        # Element2Term will add the leads automatically to fill the requested length
        box = [(0, 0), (0, 0.5), (1, 0.5), (1, -0.5), (0, -0.5), (0, 0)]
        self.segments.append(Segment(box))
        
        # Diagonal line inside the box
        self.segments.append(Segment([(0, -0.5), (1, 0)]))


# --- Drawing the Circuit ---
with schemdraw.Drawing() as d:
    d += elm.Resistor().right().label('$R_0$')
    
    d.push()
    d += elm.Line().up().length(1)
    d += (CPE := elm.CPE().right().length(5).label('$CPE_1$'))
    d += elm.Line().down().length(1)
    d.pop()
    
    d += elm.Line().down().length(1)
    d += elm.Resistor().right().length(2.5).label('$R_1$')
    d += Warburg().right().label('$W_1$')
    d += elm.Line().up().length(1)
    
```
